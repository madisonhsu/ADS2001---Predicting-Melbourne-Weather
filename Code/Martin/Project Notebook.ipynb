```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data_melb.columns = ['Year', 'Month', 'Day', 'Hour', 'Min', 'Air Temp (degrees C)',
       'Apparent Temp (degrees C)', 'Dew Pt Temp (degrees C)', 'Humidity (%)',
       'Wind Direction', 'Wind Speed (km/h)', 'Wind Gust  (km/h)',
       'MSLP (hPa)', 'Rainfall since 9 am (mm)']
```

```python
data_melb
```

```python
data_melb[['Year', 'Month', 'Day', 'Hour', 'Min']] = data_melb[['Year', 'Month', 'Day', 'Hour', 'Min']].astype(str)
```

```python
data_melb = data_melb.dropna()
```

```python
data_melb
```

```python
yearly_avg = data_melb.groupby('Year')['Air Temp (degrees C)'].mean()

yearly_avg.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Average Air Temperature (°C)')
plt.title('Average Air Temperature by Year')
plt.show()
```

```python
#The average annual temperature has remained constant over the years, but in 2025, it appears to have significantly increased. This is because the dataset only records the air temperature during the summer months of January through March. Because there are only three summer months in 2025, the average air temperature has increased.
```

```python
plt.scatter(data_melb['Wind Speed (km/h)'], data_melb['Apparent Temp (degrees C)'])
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Apparent Temperature (°C)')
plt.title('Wind Speed vs Apparent Temperature')
plt.show()
```
