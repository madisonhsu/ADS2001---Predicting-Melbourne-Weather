```python
import pandas as pd
import numpy as np

melb = pd.read_csv(
    "Melbourne01.csv",
    low_memory=False,
    on_bad_lines="skip",
    header=None
)

melb.columns = [
    "Year",
    "Month",
    "Day",
    "Hour",
    "Minute",
    "Air Temp (degrees C)",
    "Apparent Temp (degrees C)",
    "Dew Pt Temp (degrees C)",
    "Humidity (%)",
    "Wind Direction",
    "Wind Speed (km/h)",
    "Wind Gust (km/h)",
    "MSLP (hPa)",
    "Rainfall since 9 am (mm)"
]

melb["Datetime"] = pd.to_datetime(
    melb[["Year", "Month", "Day", "Hour", "Minute"]],
    errors="coerce"
)

print(melb.head())
print(melb["Datetime"].min())
print(melb["Datetime"].max())
```

```python
print(melb.head())
print(melb.columns)
print(melb["Datetime"].head())
```

```python
print(melb.shape)
print(melb.dtypes)
print(melb.isna().sum())

print(melb["Datetime"].min())
print(melb["Datetime"].max())
print(melb["Datetime"].duplicated().sum())
```

```python
melb = melb.sort_values("Datetime").reset_index(drop=True)
melb["time_diff"] = melb["Datetime"].diff()
print(melb["time_diff"].value_counts().head(10))
```

```python
num_cols = [
    "Air Temp (degrees C)",
    "Apparent Temp (degrees C)",
    "Dew Pt Temp (degrees C)",
    "Humidity (%)",
    "Wind Speed (km/h)",
    "Wind Gust (km/h)",
    "MSLP (hPa)",
    "Rainfall since 9 am (mm)"
]

melb[num_cols].describe()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(melb[col].dropna(), bins=40, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,2.5))
    sns.boxplot(x=melb[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
plt.figure(figsize=(10,4))
melb["Wind Direction"].value_counts().plot(kind="bar")
plt.title("Wind Direction Frequency")
plt.ylabel("Count")
plt.show()
```

```python
melb["Year"] = melb["Datetime"].dt.year
melb["Month"] = melb["Datetime"].dt.month
melb["Date"] = melb["Datetime"].dt.date
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return "Summer"
    elif month in [3, 4, 5]:
        return "Autumn"
    elif month in [6, 7, 8]:
        return "Winter"
    else:
        return "Spring"

melb["Season"] = melb["Month"].apply(get_season)
```

```python
monthly_mean = melb.groupby(["Year", "Month"]).agg({
    "Air Temp (degrees C)": "mean",
    "Humidity (%)": "mean",
    "MSLP (hPa)": "mean",
    "Wind Speed (km/h)": "mean"
}).reset_index()

monthly_mean["Year"] = monthly_mean["Year"].astype(int)
monthly_mean["Month"] = monthly_mean["Month"].astype(int)

monthly_mean["YearMonth"] = pd.to_datetime(
    monthly_mean["Year"].astype(str) + "-" + monthly_mean["Month"].astype(str) + "-01",
    format="%Y-%m-%d"
)

monthly_mean.head()
```

```python
plt.figure(figsize=(12,4))
plt.plot(monthly_mean["YearMonth"], monthly_mean["Air Temp (degrees C)"])
plt.title("Monthly Mean Air Temperature")
plt.xticks(rotation=45)
plt.show()
```
